```python
import dhlab.nbtext as nb
import dhlab.module_update as mu
import dhlab.graph_networkx_louvain as gnl
import dhlab.token_map as tm
from collections import Counter
from IPython.display import display, HTML, Markdown
import qgrid
```

```python
import pandas as pd

def alle_korpusnavn(korpus, ratio=0.3, cutoff=0):
    urner = nb.pure_urn(korpus)
    alle_navn = tm.combine_names(tm.corpus_names(urner, ratio = ratio, cutoff = cutoff))
    return alle_navn

def tell_navn(korpus, token_map):
    res = dict()
    for urn in korpus['urn']:
        res[urn] = tm.count_name_strings(str(urn), token_map).to_dict()[0]
    return pd.DataFrame(pd.DataFrame(res).sum(axis=1).sort_values(ascending=False))
```

# Koble URNer til gruppe

```python
name_group.keys()
```

```python
urnmap = dict()
for k in name_group:
    urnmap[k] = []
for x in korpus.itertuples(name=None):
    t = x[4].lower()
    if 'mormor' in t and not 'ellen' in t and not 'rosa' in t and not 'monrad' in t:
        urnmap['M8'].append(x[1])
    elif 'ole' in t:
        urnmap['OA'].append(x[1])
    elif 'guro' in t:
        urnmap['GU'].append(x[1])
    elif 'knerten' in t:
        urnmap['LK'].append(x[1])
    elif 'aurora' in t:
        urnmap['AU'].append(x[1])
    elif 'småtassen' in t or 'ellen' in t or 'lagertha' in t or 'monrad' in t or 'rosa' in t or 'kostemarsj' in t or 'reise' in t:
        urnmap['ES'].append(x[1])
    elif 'kaos' in x[4].lower():
        urnmap['KA'].append(x[1])
    
```
